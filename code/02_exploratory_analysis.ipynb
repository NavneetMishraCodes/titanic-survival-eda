```python
# imporing the required modules and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Plotting the column["Survived"] on a bar chart to see the no. of people died w.r.t. the no. of people survived.

```python
# data dictionary for the bar chart
survived_col_dict = titanic["Survived"].value_counts().to_dict()

for key, value in survived_col_dict.items():
    print(f"{key} : {value}")
```

```python
# plotting
plt.figure(figsize = (8, 5), dpi = 120)

plt.bar(list(reversed(survived_col_dict.keys())), list(reversed(survived_col_dict.values())), color = ["#1f77b4", "#ff7f0e"])

plt.title("Comparison Between: No. Of People Died and No. Of People Survived", fontdict = {"size" : 16, "family" : "Comic Sans MS"})
plt.xticks(ticks = [0, 1], labels = ["Survived", "Died"])
plt.xlabel("Survival Outcome", fontdict = {"size" : 15, "family" : "Comic Sans MS"})

plt.ylabel("No. Of People", fontdict = {"size" : 15, "family" : "Comic Sans MS"})

plt.show()
```

#### Plotting a bar chart showing the no. of people survived and died per class.

```python
# collecting data in a dict
survived_per_class_dict = titanic.loc[titanic["Survived"] == "Yes", "Pclass"].value_counts().sort_index().to_dict()
died_per_class_dict = titanic.loc[titanic["Survived"] == "No", "Pclass"].value_counts().sort_index().to_dict()

# printing
for key, value in survived_per_class_dict.items():
    print(f"{key} : {value}")

print("\n\n\n")

for key, value in died_per_class_dict.items():
    print(f"{key} : {value}")

```

```python
# plotting
plt.figure(figsize = (10, 4), dpi = 140)

x = np.arange(0, len(survived_per_class_dict.keys()))
w = 0.35 # width

plt.bar(x - w/2, list(survived_per_class_dict.values()), width = w, label = "Survived", color = "#1f77b4")
plt.bar(x + w/2, list(died_per_class_dict.values()), width = w, label = "Died", color = "#d62728")

plt.title("No. Of People Survived and Died per Class", fontdict = {"size" : 16, "family" : "Comic Sans MS"})
plt.xlabel("Class", fontdict = {"size" : 15, "family" : "Comic Sans MS"})
plt.ylabel("No. Of People", fontdict = {"size" : 15, "family" : "Comic Sans MS"})

plt.xticks(ticks = x, labels = ["Class 1", "Class 2", "Class 3"])
plt.legend()

plt.show()
```

```python
titanic.head(10)
```

#### Plotting a bar chart showing the *percentage* of people survived and died per class.

```python
# collecting data to plot
total_per_class = titanic["Pclass"].value_counts().sort_index().to_dict()

percent_survived_arr = (np.array(list(survived_per_class_dict.values())) * 100) / np.array(list(total_per_class.values()))

percent_died_arr = (np.array(list(died_per_class_dict.values())) * 100) / np.array(list(total_per_class.values()))

# printing 
print(percent_survived_arr)
print(percent_died_arr)
```

```python
# plotting
plt.figure(figsize = (10, 4), dpi = 140)

x = np.arange(0, len(total_per_class.keys()))
w = 0.35 # width

plt.bar(x - w/2, percent_survived_arr, width = w, label = "Survived", color = "#1f77b4")
plt.bar(x + w/2, percent_died_arr, width = w, label = "Died", color = "#d62728")

plt.title("Percentage Of People Survived and Died per Class", fontdict = {"size" : 16, "family" : "Comic Sans MS"})
plt.xlabel("Class", fontdict = {"size" : 15, "family" : "Comic Sans MS"})
plt.ylabel("People (in %)", fontdict = {"size" : 15, "family" : "Comic Sans MS"})

plt.xticks(ticks = x, labels = ["Class 1", "Class 2", "Class 3"])
plt.legend()

plt.show()
```

#### Plotting No. Of People Survived and Died per gender on a bar chart.

```python
survived_per_gender_dict = titanic.loc[titanic["Survived"] == "Yes", "Sex"].value_counts().sort_index().to_dict()
died_per_gender_dict = titanic.loc[titanic["Survived"] == "No", "Sex"].value_counts().sort_index().to_dict()

# printing
for key, value in survived_per_gender_dict.items():
    print(f"{key} : {value}")

for key, value in died_per_gender_dict.items():
    print(f"{key} : {value}")
```

```python
# plotting
plt.figure(figsize = (10, 4), dpi = 140)

x = np.arange(0, len(gender_total_dict.keys()))
w = 0.35 # width

plt.bar(x - w/2, list(survived_per_gender_dict.values()), width = w, label = "Survived", color = "#1f77b4")
plt.bar(x + w/2, list(died_per_gender_dict.values()), width = w, label = "Died", color = "#d62728")

plt.title("No. Of People Survived and Died per Gender", fontdict = {"size" : 16, "family" : "Comic Sans MS"})
plt.xlabel("Gender", fontdict = {"size" : 15, "family" : "Comic Sans MS"})
plt.ylabel("No. Of People", fontdict = {"size" : 15, "family" : "Comic Sans MS"})

plt.xticks(ticks = x, labels = ["Female", "Male"])
plt.legend()

plt.show()
```
