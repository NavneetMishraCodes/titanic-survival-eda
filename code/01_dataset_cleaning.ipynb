```python
# importing the required modules and libraries for cleaning
import pandas as pd
import numpy as np
```

#### Loading the "titanic dataset" to clean.

```python
titanic = pd.read_csv("D:\\Data_Science\\EDA_Projects\\Titanic_Dataset_EDA\\dataset\\Titanic-Dataset.csv")
```

```python
titanic.sample(10)
```

```python
titanic.info()
```

#### Renaming some columns as required.

```python
titanic.rename(columns = {"PassengerId" : "PassengerID", "Parch" : "ParCh"}, inplace = True)
```

```python
titanic.sample(10)
```

#### Dropping duplicate rows for convenience (if exists)

```python
titanic.drop_duplicates(inplace = True)
```

```python
titanic.info() # => no row was duplicate or repeated
```

#### Filling in Yes/No in the column["Survived"]

```python
titanic["Survived"] = titanic["Survived"].replace({0 : "No", 1 : "Yes"})
```

```python
titanic.sample(10)
```

#### Separating family name from the column["Name"]

```python
family_data_sr = titanic["Name"].str.split(", ").str[0] # extracting family data
```

```python
titanic["Name"] = titanic["Name"].str.split(", ").str[1] # removing the family name
```

```python
titanic.insert(4, "Family", family_data_sr) # inserting the new column
```

```python
titanic.sample(10)
```

#### Filling in M/F in the column["Sex"] for convenience.

```python
titanic["Sex"].nunique()
```

```python
titanic["Sex"] = titanic["Sex"].replace({"male": "M", "female": "F"})
```

```python
titanic.sample(10)
```

```python
titanic.groupby(["Pclass"]).agg({"Age": "mean"}) 
```

```python
titanic.groupby(["Pclass", "Sex"]).agg({"Age": "mean"}) 
```

#### Filling in the column["Age"] on the basis of columns["Pclass", "Sex"]

```python
titanic["Age"] = titanic.groupby(["Pclass", "Sex"])["Age"].transform(lambda x: x.fillna(x.median()))
```

```python
titanic.sample(10)
```

```python
titanic.info()
```

#### Replacing the column["Cabin"] with new column["Cabin_Missing"]

```python
titanic.head(10)
```

```python
titanic["Cabin_Missing"] = titanic["Cabin"].notna()
```

```python
titanic.head(10)
```

```python
# deleting
del titanic["Cabin"]
```

```python
titanic.head(10)
```

#### Filling the null values in the column["Embarked"] with the mode of it.

```python
titanic["Embarked"] = titanic["Embarked"].fillna(titanic["Embarked"].mode()[0])
```

```python
titanic.info()
```

```python
# final look
titanic.sample(10)
```

#### Saving the cleaned dataset.

```python
titanic.to_csv("D:\\Data_Science\\EDA_Projects\\Titanic_Dataset_EDA\\dataset\\Titanic-Dataset-Cleaned.csv", index = False)
```
